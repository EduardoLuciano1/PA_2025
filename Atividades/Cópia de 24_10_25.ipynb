```python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius = float(input("Digite a temperatura em Celsius: "))

fahrenheit = celsius_to_fahrenheit(celsius)

print(f"{celsius}°C é {fahrenheit}°F")
```

```python
def Calcular_potencias(n):

  quadrado = n ** 2
  cubo = n ** 3

  return quadrado, cubo

num = float(input("Escreva um número: "))

quadrado, cubo = Calcular_potencias(num)

print(f"Quadrado de {num} é {quadrado}")
print(f"Cubo de {num} é {cubo}")
```

EXERCICIO 6


```python
def calcular_desconto(valor_original, percentual_desconto):

    fator_desconto = percentual_desconto / 100

    fator_multiplicador = 1 - fator_desconto

    valor_final = valor_original * fator_multiplicador

    return valor_final



valor_inicial = float(input("Digite o valor original: "))
desconto= float(input("Digite o Percentual de Desconto (o número): "))
resultado = calcular_desconto(valor_inicial, desconto)
print(f"Produto de R$ {valor_inicial:.2f} com {desconto}% de desconto. Valor Final: R$ {resultado:.2f}")

print(f"Valor Original: R$ {valor_inicial:.2f}")
print(f"Desconto Aplicado: {desconto:.1f}%")
print(f"Valor Final a Pagar: R$ {resultado:.2f}")
```

EXERCICIO 7

```python
def encontrar_extremos(lista_de_numeros):

    maior = max(lista_de_numeros)

    menor = min(lista_de_numeros)

    return maior, menor



minha_lista = [2, 8, 9, 234, 1, 5]

maior_valor, menor_valor = encontrar_extremos(minha_lista)

print(f"Lista original: {minha_lista}")
print(f"O MAIOR número da lista é: {maior_valor}")
print(f"O MENOR número da lista é: {menor_valor}")
```

EXERCICIO 8

```python
def calculadora(a, b, operacao):


    operacao = operacao.lower()

    if operacao == 'soma':
        resultado = a + b
    elif operacao == 'subtracao':
        resultado = a - b
    elif operacao == 'multiplicacao':
        resultado = a * b
    elif operacao == 'divisao':

        if b != 0:
            resultado = a / b

        else:
            return "Erro: Divisão por zero não é permitida."
    else:
        return "Erro: Operação inválida. Use 'soma', 'subtracao', 'multiplicacao' ou 'divisao'."

    return resultado

print(f"Soma: {calculadora(10, 5, 'soma')}")

print(f"Subtração: {calculadora(20, 7, 'subtracao')}")

print(f"Multiplicação: {calculadora(4, 6, 'multiplicacao')}")

print(f"Divisão: {calculadora(100, 4, 'divisao')}")
```

EXERCICIO 10

```python
def avaliar_aluno(nota):


    if nota >= 7:
        return "Aprovado"

    elif nota >= 5 and nota < 7:
        return "Recuperação"

    else:
        return "Reprovado"

nota_digitada = float(input("Digite a nota do aluno: "))

situacao = avaliar_aluno(nota_digitada)

print(f"A nota {nota_digitada:.1f} resulta na situação: {situacao}")
```
