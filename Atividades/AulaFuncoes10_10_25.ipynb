```python
from datetime import datetime

def mostrar_hora():
  agora = datetime.now()
  print (f'O horário atual é {agora.strftime("%H:%M:%S")}')
```

```python
mostrar_hora()
```

```python
def verificar_paridade(numero):
  if numero % 2 == 0:
    print (f'O número {numero} é par.')
  else:
    print (f'O número {numero} é ímpar.')
```

```python
verificar_paridade(15)
```
