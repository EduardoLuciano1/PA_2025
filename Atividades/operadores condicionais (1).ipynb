calculadora

```python
num=int(input("digite um número: "))
op=str(input("Operação: "))
num2=int(input("digite outro número: "))
if(op=="+"):
  print(f"resultado: {num+num2}");
elif(op=="-"):
  print(f"resultado: {num-num2}");
elif(op=="*"):
  print(f"resultado: {num*num2}");
elif(op=="/"):
  print(f"resultado: {num/num2}");
```
